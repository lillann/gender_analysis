```python

from bs4 import BeautifulSoup
from os import listdir
from os.path import isfile, join

mypath = "/Users/annl/Desktop/GenieProjekt/data/Kursplaner/Raw/"
files = [join(mypath,f) for f in listdir(mypath) if isfile(join(mypath, f))]

programs_path = "/Users/annl/Desktop/GenieProjekt/data/Programdescriptions/Chalmers/"
programs_files = [join(programs_path,f) for f in listdir(programs_path) if isfile(join(programs_path, f))]
```

```python

def getItems(file,item) : 
    str  = open(file, 'r',encoding='ISO-8859-1').read()
    soup = BeautifulSoup(str,'html.parser')
    return soup.find_all(item)

    
def getContentsKey(course,val) :
    contents = course.find(attrs={"key" : val},text=True)
    if contents is None : return None
    else : return contents.get_text()
  
def getContents_program(program) : 
    return [getContentsKey(program,val) 
            for val in ["Namn_sv",
                        "Namn_en",
                        "Läsår",
                        "Utbildningsplan_sv", 
                        "Utbildningsplan_en",
                        "Syfte_sv", 
                        "Syfte_en", 
                        "Program" ]]

def getContents(course) : 
    return [getContentsKey(course,val) 
            for val in ["Namn_sv",
                        "Namn_en",
                        "Innehåll_sv",
                        "Innehåll_en",
                        "Läsår",
                        "Kurskod", 
                        "Syfte_sv", 
                        "Syfte_en", 
                        "Organisation_sv",
                        "Organisation_en",
                        "Examination_sv",
                        "Examination_sv"]]

```

```python
courses = []
for f in files  : 
    courses = courses + getItems(f,'course')
    
data = [getContents(courses[i]) for i in range(len(courses)) ]
   
```

```python

import pandas as pd

df = pd.DataFrame.from_records(data,columns=['namn_sv','namn_en','contents_sv','contents_eng','year','code','purpose_sv','purpose_eng','org_sv','org_en','ex_sv','ex_en'])
df.to_csv("kursplaner.csv")
```

```python
programs = []
for f in  programs_files :
    programs = programs + getItems(f,'pgm')
    
data = [getContents_program(programs[i]) for i in range(len(programs)) ]
```

```python


df = pd.DataFrame.from_records(data,columns=['namn_sv','namn_en','abbr','year','plan_sv','plan_en','purpose_sv','purpose_en'])
df.to_csv("programbeskrivningar.csv")
```
